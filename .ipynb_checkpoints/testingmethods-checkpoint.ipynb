```python
#if you want to clean certain data from the list
cleanfromfile = ['Alaska']
with open('USZIPCODES.txt') as oldfile, open('cleanedData.txt', 'w') as newfile:
    for line in oldfile:
        if not any(bad_word in line for bad_word in cleanfromfile):
            newfile.write(line)
```

```python
#for extracting a list of zip codes
listofzipcodes = []
with open('cleanedData.txt') as newfile:
    for line in newfile:
        splitline = line.split('	')[1]
        listofzipcodes.append(str(splitline))
        
print(listofzipcodes)
```
